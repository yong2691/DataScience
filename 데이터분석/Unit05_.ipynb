```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더는 한칸 건너뛰기 위함.

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '': #최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

print(result) #최고 기온 리스트


```

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더는 한칸 건너뛰기 위함.

result = [] # 최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] != '': #최고 기온 결측치 제외
    if row[0].split('-')[1] == '08': # 날짜를 잘라서 08월 데이터만 추출
      result.append(float(row[-1])) # 결고 리스트에 담기

plt.figure(dpi=200)
plt.plot(result,'lightpink')
plt.show()
```

매년 생일 기준 최고 기온, 평균기오느 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

high = []
low = []
avg = []

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '' :
    if 1988 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '22' :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

### 3개는 거의 필수적으로 쓰면 된다.
plt.figure(dpi=200) # 해상도 200 설정
plt.rc('font', family = 'NanumBarunGothic') # 나눔바른고딕 쓰면 한글 깨짐 없음.
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
### 3개는 거의 필수적으로 쓰면 된다.

plt.title('내 생일의 기온 변화 그래프')
plt.plot(high,'r',label='high')
plt.plot(low,'b',label='low')
plt.plot(avg,'g',label='avg')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

high = []
low = []
avg = []

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '' :
    if 1988 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '22' :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

### 3개는 거의 필수적으로 쓰면 된다.
plt.figure(dpi=200) # 해상도 200 설정
plt.rc('font', family = 'NanumBarunGothic') # 나눔바른고딕 쓰면 한글 깨짐 없음.
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
### 3개는 거의 필수적으로 쓰면 된다.

plt.title('내 생일의 기온 변화 그래프')
plt.plot(list(range(1988,2019)),high,'r',label='high')
plt.plot(list(range(1988,2019)),low,'b',label='low')
plt.plot(list(range(1988,2019)),avg,'g',label='avg')
plt.legend()
plt.show()
```
