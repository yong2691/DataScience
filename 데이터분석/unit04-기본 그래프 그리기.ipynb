```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('marker') # 타이틀 속성 추가

# 'r.--' : <색상><마커모양><선모양> 코드로 작성 가능
plt.plot([10,20,30,40],'r.-', label='circle')  # 빨간색 circle 그래프
plt.plot([40,30,20,10],'g^--', label='triangle up')  # 초록색 triangle up 그래프 linestyle -> ls 표현 가능

plt.legend(loc=6) #그래프에 범례 속성 추가
plt.show() # 그래프 보여주기
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = [] # 결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '': #최고 기온이 ''(빈문자열)이 아니면, result에다가 리스트 형태로 담아라
    result.append(float(row[-1]))

result

plt.figure(figsize=(10,2), dpi=200) #figsize(가로인치,세로인치) 1인치=2.54cm 
plt.plot(result,'b') #색상blue
plt.show()
```
