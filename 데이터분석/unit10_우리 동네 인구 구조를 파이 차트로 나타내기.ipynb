연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib as mpl
import matplotlib.pyplot as plt

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

result = [] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc ('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '지역의 남녀 성별 인구 차이')
plt.bar(range(0,101),result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]
size = [] # 인구수 사이즈 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',','')))
      size.append(int(i.replace(',',''))) #sqrt 루트를 안씌었을 때는, 너무 크게 나온다.
    for i in row[106:207]:
      f.append(int(i.replace(',','')))
    break

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

#산점도 그래프로 남녀 인구수 데이터 표현하기

plt.scatter(m,f, s=size, c=range(101), alpha=0.5, cmap = 'jet') # x축은 남자, y축은 여자
plt.plot(range(max(m)),range(max(m)),color='g')
plt.colorbar()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]
size = [] # 인구수 사이즈 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',','')))) #sqrt 루트를 씌었을때는, 제곱근으로나오니까 작게 나온다.
    for i in row[106:207]:
      f.append(int(i.replace(',','')))
    break

print(size)

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

#산점도 그래프로 남녀 인구수 데이터 표현하기

plt.scatter(m,f, s=size, c=range(101), alpha=0.5, cmap = 'jet') # x축은 남자, y축은 여자
plt.plot(range(max(m)),range(max(m)),color='g')
plt.colorbar()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]
size = [] # 인구수 사이즈 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      m.append(int(row[i]))
      f.append(int(row[i+103]))
      size.append(math.sqrt(int(row[i]) + int(row[i+103]))) #sqrt 루트를 씌었을때는, 제곱근으로나오니까 작게 나온다.
    break

print(size)

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

#산점도 그래프로 남녀 인구수 데이터 표현하기

plt.scatter(m,f, s=size, c=range(101), alpha=0.5, cmap = 'jet') # x축은 남자, y축은 여자
plt.plot(range(max(m)),range(max(m)),color='g')
plt.colorbar()
plt.show()
```
