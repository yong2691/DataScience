1월과 8월의 데이터를 히스토그램으로 그리기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv
import matplotlib.pyplot as plt


f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더는 한칸 건너뛰기 위함.

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = []
apr = []

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출

  if row[-1] != '' :
    if month == '08' : # ' ' 안에 들어있으면, string이라는 표현이다.
      aug.append(float(row[-1]))
  if row[-1] != '':
    if month == '01' :
      jan.append(float(row[-1]))
  if row[-1] != '':
    if month == '04' :
      apr.append(float(row[-1]))

#최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(aug,color= 'r', bins=100, label='aug')
plt.hist(jan,color= 'b', bins=100, label='jan')
plt.hist(apr,color= 'g', bins=100, label='apr')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

month = [ [],[],[],[],[],[],[],[],[],[],[],[]] #월별 데이터 저장할 month 변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
# 월과 같은 번호의 인덱스에 월별 데이터 저장
# (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])

plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = [] #일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([]) #day 리스트 내에 31개 리스트 생성 [ [], [], [], [], ......  31개 리스트 생성]
# 자, 31개의 그릇이 준비되었다.

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08' : #8월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1])) #-1 조심

plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(day, showfliers=False) # 아웃라이어(이상치) 값 생략
plt.show()
```
