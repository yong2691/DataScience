```python
import matplotlib.pyplot as plt
import csv
f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 입력해주세요 : ')

for row in data:
  if name in row[0] :
    for i in row[3:104] : # 남성데이터 : 3~103번 데이터 슬라이싱 - 정수데이터로 변환 - m리스트에저장
      m.append(-int(i)) # 인구수 데이터 정수로 변환 후 m 변수에 담기
    for i in row[106:]: # 여성데이터 : 106번 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i))

# 인구 구조 시각화

plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.figure(dpi=100)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name + '지역의 인구 구조')
plt.barh(range(101), m, color='b') #bar chart는 x축, y축을 무조건 넣어주어야 한다.
plt.barh(range(101), f, color='r') #bar chart는 x축, y축을 무조건 넣어주어야 한다.
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 입력해주세요 : ')

for row in data:
  if name in row[0] :
    for i in row[3:104] : # 남성데이터 : 3~103번 데이터 슬라이싱 - 정수데이터로 변환 - m리스트에저장
      m.append(-int(i.replace(',',''))) # 인구수 데이터 정수로 변환 후 m 변수에 담기
    for i in row[106:]: # 여성데이터 : 106번 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i.replace(',','')))
    break #이걸 써줘야 중복되는 것들을 다 돌지 않고, 첫번째 조건에서 스탑된다.

# 인구 구조 시각화

plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.figure(dpi=100)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name + '지역의 인구 구조')
plt.barh(range(101), m, color='b') #bar chart는 x축, y축을 무조건 넣어주어야 한다.
plt.barh(range(101), f, color='r') #bar chart는 x축, y축을 무조건 넣어주어야 한다.
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)

size = [2441,2312,1031,1233]
label = ['A형','B형','AB형','O형']
color = ['darkmagenta', 'deeppink','hotpink','pink']

plt.axis('equal')
plt.pie(size,labels=label, autopct = '%.1f%%', colors=color,explode=(0,0,0,0.2)) # 혈액형의 비율을 임의로 작성
plt.legend(loc=1)
plt.show()
# 이렇게 그리면, 누가 A형인지 누가B형인지 알수가 없으니까 label을 달아주어야 한다.
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

size = [] # 제주도의 남성 인구 합계, 여성 인구 합계
name = input('찾고 싶은 지역의 이름을 입력해주세요 : ')

for row in data:
  if name in row[0] :
    m = 0 # 남성 인구수 합계 변수
    f = 0 # 여성 인구수 합계 변수
    for i in range(101): #101번 반복할거다.
      m += int(row[3+i].replace(',','')) # m = m + int(row[3+i].replace(',',''))
      f += int(row[106+i].replace(',','')) # f = f + int(row[106+i].replace(',',''))
    break

size.append(m)
size.append(f)
print(m)
print(f)

# 파이 차트 시각화

plt.figure(dpi=100)
color = ['crimson','darkcyan']
label=['남','여']
plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%', colors = color, startangle=90, explode=(0.2,0))
plt.title (name + '지역의 남녀 성별 비율')
plt.legend()
plt.show()
```
