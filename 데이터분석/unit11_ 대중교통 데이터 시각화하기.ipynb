유임승차 / 무임승차 비율의 최댓값 찾기 -> 기존 0 데이터로 되어 있음 <br>
rate = 유임승차 인원 / 무임 승차 인원

```python
import csv


f = open('subwayfee.csv') #인코딩 utf-8 생략 가능
data = csv.reader(f)
next(data)

mx = 0 # 최애 유임승차 비율
rate = 0 # 유임승차 비율

for row in data:
  for i in range(4,8): # 4, 5, 6, 7 인덱스 데이터 정수로 변경 -> 데이터 처리 필요
    row[i] = int(row[i])
  if row[4] != 0 and row[6] != 0 and (row[4]+row[6]) > 100000 :
    rate = row[4] / (row[6]+row[4])  # 비율계산 방식 변경 rate = 유임승차 / (유임승차+무임승차)
  if rate > mx :
    mx = rate # mx에 제일 먼저 담기는 것이 서울역 87.27% 이니까, 그 뒤로는 그보다 높은애들만 나온다.
    mx_station = row[3] + ' ' + row[1] # 지하철 역 + 호선 ex) 충무로 3호선
    print(mx_station, round(mx * 100, 2),'%')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv')

data = csv.reader(f)
next(data)
next(data)

result = [] # 아침 7시 승차 데이터 추출 -> 10번째 인덱스에 존재

for row in data:
  row[4:] = map(int,row[4:])
  result.append(row[10])

result.sort() #데이터 오름차순
# 데이터 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.bar(range(len(result)), result)
plt.show()
```
