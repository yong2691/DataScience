```python
import csv
import matplotlib.pyplot as plt

f = open('pf.csv',encoding='cp949')
data = csv.reader(f)
next(data)

max_temp = []
min_temp = []
avg_temp = []

for row in data:
  if row[-1] != '':
    max_temp.append(row[-1])
  if row[-2] != '':
    min_temp.append(row[-2])
  if row[-3] != '':
    avg_temp.append(row[-3])



plt.figure(dpi=200)
plt.hist(max_temp, color='r', bins=100)
plt.hist(min_temp, color='b', bins=100)
plt.hist(avg_temp, color='g', bins=100)

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('pf.csv',encoding='cp949')
data = csv.reader(f)
next(data)

f2 = open('seoul.csv',encoding='cp949')
data2 = csv.reader(f2)
next(data2)


max_temp_D = []
max_temp_S = []

for row in data:
  if row[-1] != '':
    max_temp_D.append(float(row[-1]))

for row in data2:
  if row[-1] != '':
    max_temp_S.append(float(row[-1]))


plt.figure(dpi=200)
plt.style.use('ggplot')
plt.boxplot([max_temp_D,max_temp_S])
plt.xticks([1, 2],['daejeon', 'seoul'])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('pf.csv',encoding='cp949')
data = csv.reader(f)
next(data)

month = []
for i in range(12):
  month.append([])

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
# 월과 같은 번호의 인덱스에 월별 데이터 저장
# (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.boxplot(month, showfliers=False)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('pf.csv',encoding='cp949')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
  day.append([])

for row in data:
  if row[0].split('-')[1] == '08':
    if row[-1] != '':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))
# 일과 같은 번호의 인덱스에 일별 데이터 저장
# (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])
plt.figure(figsize=(10,2),dpi=200)
plt.style.use('ggplot')
plt.boxplot(day, showfliers=False)
plt.show()
```
