```python
# 1. pandas 선언
import pandas as pd

# 2. CSV 파일 불러오기
df = pd.read_csv('/content/세종특별자치시_공영자전거 어울링 대여소_20200915.csv',encoding='cp949')
df.head()
```

```python
df.info()
```

지도에 표시할 대여소 정보 추출

```python
df_route = df.loc[:,['STATION_NM','AREA','Y_POS','X_POS']]
df_route.head()
```

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium

map = folium.Map(location=[df_route['Y_POS'].mean(), df_route['X_POS'].mean()],
                 zoom_start=11)
for n in df_route.index:
  if df_route.loc[n, 'AREA'] == '1생활권':
    folium.Marker([df_route.loc[n,'Y_POS'],df_route.loc[n,'X_POS']],
                popup=df_route.loc[n,'STATION_NM']+' '+df_route.loc[n,'AREA'],
                icon=folium.Icon(color='green', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '2생활권':
    folium.Marker([df_route.loc[n,'Y_POS'],df_route.loc[n,'X_POS']],
                popup=df_route.loc[n,'STATION_NM']+' '+df_route.loc[n,'AREA'],
                icon=folium.Icon(color='pink', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '3생활권':
    folium.Marker([df_route.loc[n,'Y_POS'],df_route.loc[n,'X_POS']],
                popup=df_route.loc[n,'STATION_NM']+' '+df_route.loc[n,'AREA'],
                icon=folium.Icon(color='orange', icon='fa-bicycle', prefix='fa')).add_to(map)

  elif df_route.loc[n, 'AREA'] == '4생활권':
    folium.Marker([df_route.loc[n,'Y_POS'],df_route.loc[n,'X_POS']],
                popup=df_route.loc[n,'STATION_NM']+' '+df_route.loc[n,'AREA'],
                icon=folium.Icon(color='purple', icon='fa-bicycle', prefix='fa')).add_to(map)

  else:
    folium.Marker([df_route.loc[n,'Y_POS'],df_route.loc[n,'X_POS']],
                popup=df_route.loc[n,'STATION_NM']+' '+df_route.loc[n,'AREA'],
                icon=folium.Icon(color='red', icon='fa-bicycle', prefix='fa')).add_to(map)

map
```

```python
# 지도에 표시하기 위해 folium 라이브러리
import folium
map = folium.Map(location=[df_route['Y_POS'].mean(), df_route['X_POS'].mean()],
                           zoom_start = 11)
c = ['green','blue','red','gray','pink','black']
for n in df_route.index:
    for i in range(0,5):
        if str(df_route.loc[n, 'AREA'][0]) == str(i+1):
            folium.Marker([df_route.loc[n, 'Y_POS'], df_route.loc[n, 'X_POS']], # 위도, 경도
            popup = df_route.loc[n, 'STATION_NM'] + ' ' + df_route.loc[n, 'AREA'],
            icon=folium.Icon(color=c[i], icon='fa-bicycle', prefix='fa')).add_to(map)
            break  
        else:
            folium.Marker([df_route.loc[n, 'Y_POS'], df_route.loc[n, 'X_POS']], # 위도, 경도
            popup = df_route.loc[n, 'STATION_NM'] + ' ' + df_route.loc[n, 'AREA'],
            icon=folium.Icon(color='orange', icon='fa-bicycle', prefix='fa')).add_to(map)
map
```

CCTV 공공데이터 준비하기

```python
# CSV 파일 불러오기
cctv_df = pd.read_csv('/content/세종특별자치시_CCTV_20210101.csv',encoding='cp949')
cctv_df.head()
```

```python
#지도에 표시할 CCTV 정보 추출
cctv_route = cctv_df.loc[:,['카메라화소수','위도','경도']]
cctv_route.head()
```

```python
#지도 시각화 - 지도에 정보 표출 위해 folium 라이브러리, Marker 활용하여 지도 표출
import folium

for n in cctv_route.index[:100]:
  folium.Marker([cctv_route.loc[n,'위도'],
                 cctv_route.loc[n,'경도']],
                  icon=folium.Icon(icon='video-camera', prefix='fa')).add_to(map)
map
```

```python
#전체 CCTV만 따로 지도에 표시해보기
import folium

cctv_ndf = pd.read_csv('/content/세종특별자치시_CCTV_20210101.csv',encoding='cp949')

cctv_route = cctv_ndf.loc[:,['카메라화소수','위도','경도']]

cctv_map = folium.Map(location=[cctv_route['위도'].mean(), cctv_route['경도'].mean()],
                                zoom_start=11)

for n in cctv_route.index:
  folium.Marker([cctv_route.loc[n,'위도'],
                 cctv_route.loc[n,'경도']],
                  icon=folium.Icon(icon='video-camera', prefix='fa')).add_to(cctv_map)
cctv_map #NaN값 처리 해야 한다.
```
