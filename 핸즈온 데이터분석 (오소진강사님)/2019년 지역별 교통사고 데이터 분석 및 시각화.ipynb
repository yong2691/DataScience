```python
#라이브러리 임포트
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('도로교통공단_교통사고 정보.csv',encoding='euc-kr')
df.head()
```

```python
df.info()
```

교통사고 시간 전처리

```python
#Substr 문자열 자르기를 위해 데이터타입 변경
df = df.astype({'발생년월일시' : 'string'}) #데이터프레임에서 데이터타입을 변경할 수 있다.
```

```python
df.info()
```

```python
#2019010100 -> 00 시간으로 분리
df['발생시간'] = df['발생년월일시'].str[8:] # 발생시간 이라는 컬럼이 만들기
df.info() 
```

```python
#발생시간을 숫자 데이터 타입으로 변경
df = df.astype({'발생시간' : 'int64'})
df.info()
```

```python
# 2019010100 -> 20190101 추출 -> 2019-01-01 (날짜형태 YYYY-MM-DD 형태로 변경)
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8], format='%Y-%m-%d', errors='raise')
```

```python
df.info()
```

```python
df.head()
```

```python
fig = px.scatter(df,x="발생년월일시", y="발생시간", color='발생지시도', hover_data=['발생시간'],size='발생시간')
fig.show()
# hover_data=['발생시간'] 
# size=['발생시간'] 발생시간에 따라 버블의 사이즈가 달라진다.
```

시간대 별 교통사고 사망자 현황

```python
fig = px.bar(df, x='사망자수', y='발생시간', orientation='h') # orientation='h' 은 수평으로그리기
fig.show()
```

시도 별 교통사고 사망자 현황

```python
fig=px.bar(df, x='사망자수',y='발생지시도', orientation='h', title='시도별 교통사고 사망자 현황')
fig.show()
#이 데이터의 허점은? 면적이나 인구를 고려하지 않았다.
# 제대로 보려면, 사망자 / 총인구수 으로 나누어서 비율을 보아야 되겠다.
```

지도 활용하여 교통사고 현황 분석

```python
import folium
map = folium.Map(location=[36.48022677322587, 127.28883989950721])
```

```python
map
```

```python
df_map = df[df.발생지시도=="세종"]
df_map.info()
```

```python
df_map
```

사망자수, 부상자수, 중상자수, 경상자수 전처리 -> 지도에 원을 그리기 위해서

```python
df_map = df_map.astype({'사망자수' : 'float64'})
df_map = df_map.astype({'경상자수' : 'float64'})
df_map = df_map.astype({'중상자수' : 'float64'})
df_map = df_map.astype({'부상자수' : 'float64'})
```

```python
df_map.info()
```

사망자수, 경상자 수에 따른 CircleMarker 지도에 표출하기

```python
for n in df_map.index:
  cnt = df_map['사망자수'][n] + df_map['경상자수'][n] #사망자수 + 경상자수
  folium.CircleMarker([df_map['위도'][n], df_map['경도'][n]],
                       radius=cnt*10, popup=df_map['사고유형'][n]+df_map['가해자_당사자종별'][n]+df_map['피해자_당사자종별'][n], color='#3186cc',
                       fill_color='#3186cc').add_to(map)
map.save('세종교통사고현황.html')
```
