항아리 그래프를 위해 남성 데이터 음수로 바꿔보기

```python
import matplotlib.pyplot as plt
import csv
f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

#for문 두개로 돌리기
for row in data:
  if '신도림' in row[0] :
    for i in row[3:104] : # 남성데이터 : 3~103번 데이터 슬라이싱 - 정수데이터로 변환 - m리스트에저장
      m.append(-int(i)) # 인구수 데이터 정수로 변환 후 m 변수에 담기
    for i in row[106:]: # 여성데이터 : 106번 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i))

print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=200)
plt.rc ('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh (range(101), m,color='b',label='남성')
plt.barh (range(101), f,color='r',label='여성')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

#for문 두개로 돌리기
for row in data:
  if '새롬동' in row[0] :
    for i in row[3:104] : # 남성데이터 : 3~103번 데이터 슬라이싱 - 정수데이터로 변환 - m리스트에저장
      m.append(-int(i)) # 인구수 데이터 정수로 변환 후 m 변수에 담기
    for i in row[106:]: # 여성데이터 : 106번 ~ 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f리스트에 저장
      f.append(int(i))

plt.figure(dpi=200)
plt.rc ('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('새롬동 지역의 남녀 성별 인구 분포')
plt.barh (range(101), m,color='b',label='남성')
plt.barh (range(101), f,color='r',label='여성')
plt.legend()
plt.show()
```
