```python
import csv

max_temp = -999 # 최고 기온 값을 저장할 변수 ,     -999의 의미는? 초기값로서 인식하게 하기 위함. 0으로 넣어도 된다. 근데 혹시 최고기온이 0이 될수도 있으니까 -999로 씀. 나중에 최고값이 0이 떴을 때, 이게 진짜 최고온도인지 아님 오류가 나서 0이 나온건지 알수가 없기 때문이다!
max_date = ''  #최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open("seoul.csv",'r', encoding='cp949')

data = csv.reader(f)
header = next(data) # 헤더 추출

for row in data:

  #최고 기온 데이터가 ''(결측치) 확인
  if row[-1] == '': #최고기온데이터가 결측치라면,
    row[-1] = -999 # -999를 넣어서 빈 문자열('')이 있던 자리라고 표시
  row[-1] = float(row[-1]) # 기온 데이터와 실수형으로 변환

  #최고 기온 정보를 저장

  if max_temp < row[-1]:      # 최고 기온이 -999보다 크다면?
    max_date = row[0]         # 최고 기온이 나온날의 date 를 담아라
    max_temp = row[-1]        # 최고 기온이 나온날의 온도 를 담아라

f.close()

print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은' , max_date,'로', max_temp, '도 였습니다.')
```

step 4) 최저 기온과 최저 기온이었던 날짜 찾기 (스스로 해보기)

```python
import csv

min_temp = 999 # 최고 기온 값을 저장할 변수 -> 명확한 결과를 위해 극적인 값으로 셋팅
min_data = '' # 최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open('seoul.csv','r',encoding='cp949')
data =csv.reader(f) 
header = next(data) #헤더 추출

for row in data :
  #최저 기온 데이터가 ''(결측치) 확인 / 결측치가 있으면 문제가 생긴다.
  if row[-2] =='':
    row[-2] = 999 # 999을 넣어서 빈 문자열('')이 있던 자리라고 표시 이것을 하는 이유는? 결측치가 있으면 float 타입으로 변환하라는 부분에서 오류가 난다. 최저기온은 999라고 해준이유는? 절대 최저기온이 999는 안나올거아냐? 그러니까 999로 넣은것이다. -999로 넣으면 -999보다 작은값은 없을테니까,
  row[-2] = float(row[-2])
  # 최저 기온 정보를 저장

  if min_temp > row[-2]: 
    min_temp = row[-2]
    min_date = row[0]
f.close()    

print("기상 관측이래 서울의 최저 기온이 가장 높았던 날은", min_date,"로",min_temp,"도 였습니다.")
print(f"기상 관측이래 서울의 최저 기온이 가장 높았던 날은 {min_date}로 {min_temp}도 였습니다")
print("기상 관측이래 서울의 최저 기온이 가장 높았던 날은 {}로 {}도 였습니다.".format(min_date,min_temp))
```
