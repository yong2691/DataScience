```python
import csv

f= open("/content/seoul.csv", 'r', encoding='cp949')
data = csv.reader(f,delimiter=',') #delimiter(구분자) -> , 로 생략가능 delimiter 제일많이 쓰는 순서 1. ,(컴마) 2. ^(캐럿) 3. |(파이프)
# 데이터를 한줄씩 출력

for row in data:
  print(row)
f.close()
#이렇게 하면 자료의 처음 시작 형태는 리스트 [ ] 형태로 담아지게 된다.
```

```python
import csv
f = open('seoul.csv',encoding='cp949')
data = csv.reader(f)
header = next(data) #header라는 변수에 next라는 함수를 넣는다. next는 첫번째 행을 읽은 후 다음 행으로 이동 시키는 함수
print(header)
print(header)
print(next(data))
f.close()
```

```python
import csv

f= open("/content/seoul.csv", 'r', encoding='cp949') #파일을 읽는것을 변수로 넣고

data = csv.reader(f) # csv라이브러리에서 reader라는 함수를 써서 파일을 여는 변수를 넣으면 데이터가 열리고
header = next(data) #헤더를 추출
print(header)

# 데이터를 한줄씩 출력(추출)

for row in data:
  print(row)

f.close()
```
