```python
import csv 

max_temp = -999 #최고 기온 값을 저장할 변수
max_data = '' #최고 기온이 가장 높았던 날짜를 저장하는 변수

f= open('/content/pf.csv','r',encoding ='cp949')
data = csv.reader(f)
header = next(data) #헤더 추출

for row in data:

  #최고 기온 데이터에 ''(결측치) 확인
  if row[-1] == '':
    row[-1] = -999
  #데이터 실수형으로 변환
  row[-1]=float(row[-1])

  #최고 기온 정보를 저장
  if max_temp < row[-1]:
    max_data = row[0]
    max_temp = row[-1]
f.close()

print('교수님이 태어난 이래로 대전의 최고 기온이 가장 높았던 날은 ',max_data,'로',
      max_temp,'도 였습니다.')
```

```python
import csv 

min_temp = 999 #최저 기온 값을 저장할 변수
min_data = '' #최저 기온이 가장 높았던 날짜를 저장하는 변수

f= open('/content/pf.csv','r',encoding ='cp949')
data = csv.reader(f)
header = next(data) #헤더 추출

for row in data:

  #최저 기온 데이터에 ''(결측치) 확인
  if row[-2] == '':
    row[-2] = 999
  #데이터 실수형으로 변환
  row[-2]=float(row[-2])

  #최저 기온 정보를 저장
  if min_temp > row[-2]:
    min_data = row[0]
    min_temp = row[-2]
f.close()

print('교수님이 태어난 이래로 대전의 최저 기온이 가장 낮았던 날은 ',min_data,'로',
      min_temp,'도 였습니다.')
```

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

f = open('pf.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

result = [] # 결측치를 제거한 최고 기온 리스트
date = []

for row in data:
  if row[-1] != '': #최고 기온이 ''(빈문자열)이 아니면, result에다가 리스트 형태로 담아라
    result.append(float(row[-1]))
    date.append(row[0])  

plt.figure(figsize=(50,6), dpi=200) # dpi는 해상도
plt.plot(date,result,'b',label='Max_Temp') #색상blue
plt.xlabel('date')
plt.ylabel('temp')
plt.legend(loc=5)
plt.show()
```
