연령 별 인구수 (0~100세 이상) 데이터 확인 -> 4번

```python
import csv
f = open('age.csv', encoding='cp949')
data = csv.reader(f)

for row in data:
  if '신도림' in row[0]:
    print(row[3:]) # 0~100세 이상 인구수 출력, row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수
    print(len(row[3:])) # 0~100세 이상 데이터 수 출력 -> 101개
```

우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
import csv
f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

for row in data:
  if '신도림' in row[0]:
    result.append(row[3:]) # 0~100세 이상 인구수 출력, row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수

print(result)
#이렇게하면 문제가, 문자형으로 들어오게 된다. 그럼 result.append(int(row[3:])) 으로하면 안되나?
# int를 씌우면 오류가 난다.
```

```python
import csv
f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i)) # 인구수 데이터 변환 후, result 변수에 담기

print(result)
```

우리 동네 연령별 인구수 데이터 시각화하기

```python
import matplotlib.pyplot as plt
import csv
f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:]:
      result.append(int(i))
  
plt.figure(dpi=200)
plt.style.use('ggplot') # 격자 무늬 스타일로 지정 classic, dark_background
plt.plot(result)
plt.show()
```
