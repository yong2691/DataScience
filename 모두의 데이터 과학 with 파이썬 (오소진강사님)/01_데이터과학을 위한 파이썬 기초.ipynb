```python
from collections import Counter

phrase = "a man a plan a canal panama"
ls = phrase.split()
ls
```

```python
cntr = Counter(ls)
cntr
```

```python
cntr.most_common(1) #가장 빈번하게 등장한 n개의 아이템과 빈도 추출
```

```python
type(cntr) #cntr 은 컬렉션 모듈에 들어있는 카운터 타입이다.
```

```python
cntr.most_common(2) #man이 먼저 나온 이유는? 동점일때, 리스트에서 순서로 가져온다. 
```

```python
cntr.most_common(5) # 순서대로 5개까지 나와라.
```

```python
cntr.most_common(6) # 순서대로 6개까지 나와라. 근데 총 5개니까 5개로 끝난다.
```

```python
cntr.most_common() # 소괄호() 안에 아무것도 안쓰면
```

```python
#위에것이 딕셔너리처럼 보이지만, 딕셔너리형태처럼 보이는 카운터이다.
#위에것은  [ ] 대괄호로 둘러쌓여 있으니까 리스트이다.

cntrDict = dict(cntr)
cntrDict
#이렇게 { } 으로 둘러쌓여 있어야 딕셔너리다.
```

```python
#search 문자열 일부분이 정규식과 매칭되는지 확인 (매칭되는게 앞부분에 없다면 이 함수 사용)
# [a-z] : 소문자들만 일치
re.search(r"[a-z]","0010010 Does not 010 Starts with a number").group()
```

```python
#search 문자열 일부분이 정규식과 매칭되는지 확인 (매칭되는게 앞부분에 없다면 이 함수 사용)
# [a-z] : 소문자들만 일치
re.search(r"[a-z]+","0010010 Does not 010 Starts with a number").group()
```

```python
#search 문자열 일부분이 정규식과 매칭되는지 확인 (매칭되는게 앞부분에 없다면 이 함수 사용)
# [a-z] : 소문자들만 일치
# re.I : 대소문자 구분 하지 않음.
re.search(r"[a-z]+","0010010 Does not 010 Starts with a number", re.I).group()
```

```python
#search 문자열 일부분이 정규식과 매칭되는지 확인 (매칭되는게 앞부분에 없다면 이 함수 사용)
# [A-Z] : 대문자들만 일치
re.search(r"[A-Z]+","0010010 Does not 010 Starts with a number", re.I).group()
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기(대소문자 무시) -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"[A-z]+", "0010010 Does not 010 Starts with a number", re.I)
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기(대소문자 무시) -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"\d+", "0010010 Does not 010 Starts with a number", re.I)
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기(대소문자 무시) -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"\D+", "0010010 Does not 010 Starts with a number")
```

```python
# findall : 정규식에 부합하는 모든 문자열 찾기(대소문자 무시) -> 모든 소문자 들어 있는 토큰 찾기
re.findall(r"[^a-z]+", "0010010 Has at least one 010 letter 0010010")
```

```python
# sub : string으로 매칭되는 부분 문자열을 repl로 치환
re.sub(r"[a-z]+", '[...]', "0010010 Has at least one 010 letter 0010010", flags = re.I)
# flags는? 대소문자 구분하지 않고 쓰는것을 뜻함. sub함수에는 flags=re.I 를 쓴다.
# 또는 re.I 앞에 0, 를 써줘도 된다.
```

```python
# sub : string으로 매칭되는 부분 문자열을 repl로 치환
re.sub(r"[a-z]+", '[...]', "0010010 Has at least one 010 letter 0010010", 0, re.I)
# flags는? 대소문자 구분하지 않고 쓰는것을 뜻함. sub함수에는 flags=re.I 를 쓴다.
# 또는 re.I 앞에 0, 를 써줘도 된다.
```
